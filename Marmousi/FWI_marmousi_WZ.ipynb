```python
import os
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1 import make_axes_locatable
from utils_denise import ModLoader, PltModel
```

```python
datadir = '../model'
basename = 'modelTest_{}_stage_4.bin'
nx = 500
nz = 174
dx = 20.
mod = ModLoader(datadir, basename, nx, nz, dx)
cmap = mod.read_cmap('.', 'cmap_cm.pkl')
```

```python
PltImg = PltModel((mod.vp/1000, mod.vs/1000, mod.rho/1000),
                  ('vp', 'vs', 'rho'), ('km/s', 'km/s', 'g/cc'), 0,
                  mod.height, mod.width, ytitle='Depth')
fig = PltImg.viewMulti([0, 1, 2], (3, 1), (6, 2), cmap=cmap)
plt.show()
```
